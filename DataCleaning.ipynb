Objectif : clean les données

```python
# Libraries
from sklearn import decomposition
import pandas as pd
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('kidney_disease.csv',sep=',')
N = data.shape[1]

data.head()
```

```python
data[['htn','dm','cad','pe','ane']] = data[['htn','dm','cad','pe','ane']].replace(to_replace={'yes':1,'no':0})
data[['rbc','pc']] = data[['rbc','pc']].replace(to_replace={'abnormal':1,'normal':0})
data[['pcc','ba']] = data[['pcc','ba']].replace(to_replace={'present':1,'notpresent':0})
data[['appet']] = data[['appet']].replace(to_replace={'good':1,'poor':0,'no':np.nan})
data['classification'] = data['classification'].replace(to_replace={'ckd':1,'notckd':0})
```

```python
data.mean()
```

```python
column_float = ['age','bp','sg','al','su','bgr','bu','sc','sod','pot','hemo']

for col in column_float:
    data[col]=data[col].fillna(data.mean()[col]) 

column_string = ['rbc','pc','pcc','ba','htn','appet','pe','ane']
for col in column_string:
    if (data.mean()[col]<0.5):
        data[col]=data[col].fillna(0) 
    else:
        data[col]=data[col].fillna(1) 
```
